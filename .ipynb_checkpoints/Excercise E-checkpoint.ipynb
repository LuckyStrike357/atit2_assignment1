```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import f1_score
import scipy.spatial.distance
import seaborn as sns
from sklearn.manifold import TSNE

plt.gray()
import skimage as ski
```

```python
cifar = tf.keras.datasets.cifar10.load_data()[0] 
X_train, Y_train = cifar[0], cifar[1]
X_test, Y_test = cifar[0], cifar[1]
```

```python
#convert CIFAR 10 to grayscale
X_train = ski.color.rgb2gray(X_train)
X_test = ski.color.rgb2gray(X_test)
```

```python
#flatten data
n_train = X_train.shape[0]
n_test = X_test.shape[0]

m = X_train.shape[1]**2

X_train = X_train.reshape([n_train, m])
X_test = X_test.reshape([n_test, m])
```

```python
X_train = X_train.reshape([n_train, m])
X_test = X_test.reshape([n_test, m])

idx = np.random.randint(n_train, size=10000)
x_train_sample = X_train[idx]
y_train_sample = Y_train[idx]

idx = np.random.randint(n_test, size=100)
x_test_sample = X_test[idx]
y_test_sample = Y_test[idx]
```

# KNN and Cifar

```python
from sklearn.neighbors import KNeighborsClassifier
knn_2 = KNeighborsClassifier(n_neighbors=2)
knn_4 = KNeighborsClassifier(n_neighbors=4)
knn_8 = KNeighborsClassifier(n_neighbors=8)
```

```python
knn_2.fit(x_train_sample, np.ravel(y_train_sample))
knn_4.fit(x_train_sample, np.ravel(y_train_sample))
knn_8.fit(x_train_sample, np.ravel(y_train_sample))
```

```python
# Predict the results
knn_2_predictions = knn_2.predict(x_test_sample)
knn_4_predictions = knn_4.predict(x_test_sample)
knn_8_predictions = knn_8.predict(x_test_sample)
```

```python
print("KNN 2 predictions: {}".format(knn_2_predictions[:5]))
print("KNN 4 predictions: {}".format(knn_4_predictions[:5]))
print("KNN 8 predictions: {}".format(knn_8_predictions[:5]))
```

```python
# The score method uses the f1_score to calculate
knn_2_score = knn_2.score(x_test_sample, y_test_sample)
knn_4_score = knn_4.score(x_test_sample, y_test_sample)
knn_8_score = knn_8.score(x_test_sample, y_test_sample)
print("KNN 2 score: {}".format(knn_2_score))
print("KNN 4 score: {}".format(knn_4_score))
print("KNN 8 score: {}".format(knn_8_score))
```

```python
# Get those pictures that were misclassified
def get_misclassified_pictures(predictions):
    index = 0
    misclassifiedIndexes = []
    for label, predict in zip(Y_train, predictions):
        if label != predict: 
            misclassifiedIndexes.append(index)
        index +=1
    return misclassifiedIndexes
```

```python
def show_misclassified_pictures(predictions, misclassifiedIndexes):
    plt.figure(figsize=(20,4))
    for plotIndex, badIndex in enumerate(misclassifiedIndexes[0:5]):
        plt.subplot(1, 5, plotIndex + 1)
        plt.imshow(np.reshape(X_train[badIndex], (32,32)), cmap=plt.cm.gray)
        plt.title('Predicted: {}, Actual: {}'.format(predictions[badIndex], Y_train[badIndex,0]), fontsize = 15)
        
        
knn_2_indexes = get_misclassified_pictures(knn_2_predictions)
knn_4_indexes = get_misclassified_pictures(knn_4_predictions)
knn_8_indexes = get_misclassified_pictures(knn_8_predictions)

show_misclassified_pictures(knn_2_predictions, knn_2_indexes)
show_misclassified_pictures(knn_4_predictions, knn_4_indexes)
show_misclassified_pictures(knn_8_predictions, knn_8_indexes)
```

```python
#pd.DataFrame(np.array(score).T, columns=["k = 2", "k = 4", "k = 8"])
from sklearn.metrics import classification_report
print(classification_report(y_test_sample, knn_2_predictions))
print(classification_report(y_test_sample, knn_4_predictions))
print(classification_report(y_test_sample, knn_8_predictions))
```

As we can see KNN works very badly with this converted Dataset. 

# KMeans and Cifar

```python
from sklearn.cluster import KMeans
kmeans_10 = KMeans(n_clusters=10, max_iter=10).fit(x_train_sample)
kmeans_100 = KMeans(n_clusters=10, max_iter=100).fit(x_train_sample)
kmeans_1000 = KMeans(n_clusters=10, max_iter=1000).fit(x_train_sample)
```

```python
def calculate_scores(model):
    scores = []
    count = 0
    
    for i in range(10):
        scores.append([])
    
    for value in model.labels_:
        scores[value].append(Y_train[count])
        count += 1
        
    return scores

kmeans_10_scores = calculate_scores(kmeans_10)
kmeans_100_scores = calculate_scores(kmeans_100)
kmeans_1000_scores = calculate_scores(kmeans_1000)
```

```python
kmeans_10_clusters = []
kmeans_100_clusters = []
kmeans_1000_clusters = []
local_array = []

for i in range(10):
    # 10 iterations    
    local_array = np.bincount(np.ravel(kmeans_10_scores[i]))
    kmeans_10_clusters.append(local_array)
    # 100 iterations
    local_array = np.bincount(np.ravel(kmeans_100_scores[i]))
    kmeans_100_clusters.append(local_array)
    #1000 iterations
    local_array = np.bincount(np.ravel(kmeans_1000_scores[i]))
    kmeans_1000_clusters.append(local_array)
```

```python
def calc_cluster_percentage(cluster):

    percentages = []
    
    for array in cluster:
        max_idx = np.where(array == max(array))[0]
        if len(max_idx) == 1:
            percentages.append(max(array) / sum(array))
        else:
            percentages.append(-1)

    return(percentages)

def get_majority_labels(cluster):
    
    labels = []
    
    for array in cluster:
        max_idx = np.where(array == max(array))[0]
        if len(max_idx) == 1:    
            labels.append(max_idx[0])
        else:
            labels.append(-1)
    
    return labels
    
def get_missing_labels(labels):
    classes = np.arange(10)
    return np.where(np.isin(classes, labels) == False)[0]

cluster_10_perc = calc_cluster_percentage(kmeans_10_clusters)
cluster_10_labels = get_majority_labels(kmeans_10_clusters)

cluster_100_perc = calc_cluster_percentage(kmeans_100_clusters)
cluster_100_labels = get_majority_labels(kmeans_100_clusters)

cluster_1000_perc = calc_cluster_percentage(kmeans_1000_clusters)
cluster_1000_labels = get_majority_labels(kmeans_1000_clusters)
```

```python
cluster_100_perc
cluster_10_labels
print("Missing labels for 10 iterations:" , get_missing_labels(cluster_10_labels))
```

```python
combined_percentages = [cluster_10_perc, cluster_100_perc, cluster_1000_perc]
df1 = pd.DataFrame(np.array(combined_percentages).T, columns=["10","100","1000"])
np.array(combined_percentages).T
df1
```
